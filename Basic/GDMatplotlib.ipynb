```python
'''apply styles'''
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [5,3,4,5,6]
y = [1,2,3,7,8]

x1 = [1,3,4,5,6]
y1 = [1,2,3,9,10]

plt.plot(x,y, 'g', linewidth = 5)
plt.plot(x1,y1, 'c', linewidth = 10)
plt.title('Test chart')
plt.xlabel('somthing-x')
plt.ylabel('somthing-y')
plt.show()
```

```python
'''apply styles'''
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')

x = [5,3,4,5,6]
y = [1,2,3,7,8]

x1 = [1,3,4,5,6]
y1 = [1,2,3,9,10]

plt.plot(x,y, 'g', linewidth = 5, label = 'line1')
plt.plot(x1,y1, 'c', linewidth = 10, label = 'line2')
plt.title('Test chart')
plt.xlabel('somthing-x')
plt.ylabel('somthing-y')
plt.legend()
plt.grid(True,color='b')

plt.show()
```

Scatter Char

```python
from matplotlib import pyplot as plt
from matplotlib import style

#style.use('ggplot')

x = [5,3,4,5,6]
y = [1,2,3,7,8]

x1 = [1,3,4,5,6]
y1 = [1,2,3,9,10]

plt.scatter(x,y, label = 'line1', color = 'g')
plt.scatter(x1,y1, label = 'line2', color = 'r')

plt.title('Test chart')
plt.xlabel('somthing-x')
plt.ylabel('somthing-y')

plt.show()
```

Bar Chart

```python
from matplotlib import pyplot as plt
from matplotlib import style

#style.use('ggplot')

x = [1,2,3,5,6]
y = [1,2,3,7,8]

x1 = [1,3,4,5,6]
y1 = [1,2,3,9,10]

plt.bar(x,y,color = 'g', align = 'center')
plt.bar(x1,y1,color = 'r', align = 'center')

plt.title('Test chart')
plt.xlabel('somthing-x')
plt.ylabel('somthing-y')

plt.show()
```

Read from CSV to Plot

```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np

style.use('ggplot')

x,y = np.loadtxt('example.csv', delimiter = ',', unpack = True)

plt.plot(x,y)
plt.title('example.csv')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
