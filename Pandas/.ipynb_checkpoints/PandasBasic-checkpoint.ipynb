```python
import pandas as pd
import quandl
import pickle
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

#df = quandl.get('FMAC/HPI_TX',authtoken="***")

authtoken_api = '***'
fiddy_states = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states')

def state_list():
    return fiddy_states[0][1][1:]


def grab_initial_state_data():
    
    main_df = pd.DataFrame()

    for abbv in state_list():
        query = 'FMAC/HPI_'+str(abbv)
        #print(query)
        df = quandl.get(query, authtoken = authtoken_api)
        #to_get the percent changes
        df = df.pct_change()
        
        df.columns = [abbv]
        #print(df.head())
        if main_df.empty:
            main_df = df
        else:
            main_df = main_df.join(df)
                
        #serialize and store the result into the pickle object 
        pickle_out = open('fidddy_states.pickle','wb')
        pickle.dump(main_df, pickle_out)
        pickle_out.close()


grab_initial_state_data()
        
#read HPI data.
HPI_data = pd.read_pickle('fidddy_states.pickle')

#changing the column
#HPI_data['TX2'] = HPI_data['TX'] * 2
#print(HPI_data[['TX2','TX']])

HPI_data.plot()
plt.legend().remove()
plt.show()
```

Percent change from the begining 

```python
import pandas as pd
import quandl
import pickle
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

#df = quandl.get('FMAC/HPI_TX',authtoken="***")

authtoken_api = '***'
fiddy_states = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states')

def state_list():
    return fiddy_states[0][1][1:]


def grab_initial_state_data():
    
    main_df = pd.DataFrame()

    for abbv in state_list():
        query = 'FMAC/HPI_'+str(abbv)
        df = quandl.get(query, authtoken = authtoken_api)
        df.columns = [abbv]
        
        df[abbv] = (df[abbv] - df[abbv][0]) / df[abbv][0] * 100
        
        if main_df.empty:
            main_df = df
        else:
            main_df = main_df.join(df)
                
        #serialize and store the result into the pickle object 
        pickle_out = open('fidddy_states2.pickle','wb')
        pickle.dump(main_df, pickle_out)
        pickle_out.close()

grab_initial_state_data()
        
#read HPI data.
HPI_data = pd.read_pickle('fidddy_states2.pickle')

#changing the column
#HPI_data['TX2'] = HPI_data['TX'] * 2
#print(HPI_data[['TX2','TX']])

HPI_data.plot()
plt.legend().remove()
plt.show()
```

```python
def HPI_benchmark():
    df = quandl.get('FMAC/HPI_USA', authtoken = authtoken_api)
    df.columns = ['USA']
    df['USA'] = (df['USA'] - df['USA'][0]) / df['USA'][0] * 100.0
    return df
    
    
fig = plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))

HPI_data = pd.read_pickle('fidddy_states2.pickle')
benchmark = HPI_benchmark()

HPI_data.plot(ax = ax1)
benchmark.plot(ax = ax1, color = 'k', linewidth=10)
    
plt.legend().remove()
plt.show()
```
